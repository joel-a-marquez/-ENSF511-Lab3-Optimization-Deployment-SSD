##### The results from resizing the images from 224x224 to 112x112

```python
# Libraries

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.DataFrame({
    "image_size" : ["(224x224)","(112x112)"],
    "accuracy" : [96.875,94.271]
})
```

```python
plt.figure(figsize=(9, 7), dpi=100)
ax = sns.barplot(y=df.accuracy, x = df.image_size)
ax.set_ylim([0, 100])
ax.set_ylabel("Accuracy (%)")

# Iterrating over the bars one-by-one
for bar in ax.patches:
    
    plt.annotate(format(bar.get_height(), '.3f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()-5), ha='center', va='center',
                   size=15, xytext=(0, 8),
                   textcoords='offset points')
```

**Figure 1. Accuracy vs. Input Image Size**

Figure 1 above shows the results of training the images of two different sizes. The image size of (224x224) and (112x112) had the accuracy (evaluated with the test_dataset) of 96.875% and 94.271% respectively. This suggests that reducing the image size will decrease the accuracy of the model. Less features to extract from the model (less resolution of the image) will mean that it will slightly affect the ability to distinguish one label to another.
